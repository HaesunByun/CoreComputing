## 추상 데이터 타입의 개념

전반부에서는 어떤 문제를 해결하고자 할때
순차적, 조건적, 반복적으로 생각하는 연습을 했고, 재귀를 통해 생각의 전환으로 문제를 해결하는 연습을 했다.
이것은 주로 프로세스에 관한 관점이었다. 문제를 해결하는 방법에 있어서 흐름(flow), 즉 이를 **프로세스**라고 표현하는데 어떤 방식으로 흘러가면서 문제를 해결해야 할지에 대해 순서적 관점으로 생각하는 것이다.


이러한 프로세스를 순서적, 절차적으로 바라보지 않고 다르게 바라보는 방법으로 클래스와 객체에 대해서도 다루었다.

어떤 문제를 해결하기 위한 방법으로 프로세스뿐만 아니라 여러 가지 중요한 팩터들이 있다. 프로세스를 어떻게 디자인할 것인가도 중요한 팩터이지만 또 다른 관점으로 데이터를 어떻게 정의하고 조작할 것인가도 하나의 중요한 팩터이다.

이번 장에서는 데이터를 어떻게 다룰 것인가의 관점에서 집중적으로 다뤄보고자 한다.


지금까지 우리는 int, float, str, list, dict, tuple, set 등 다양한 데이터의 타입을 다뤄보았다. 이런 타입들은 원래부터 본질적으로 존재했던 것은 아니다. 실제 컴퓨터의 기계레벨에서는 0과 1 타입만 존재한다. 그 외의 타입들은 0과 1위에 덧붙여져서 만들어진 파생된(drived) 타입들이다.

주로 프로그래밍 언어를 디자인하는 사람들이 설계해서 만들어 놓았고, 여러분들이 설계해서 만들어 놓을 수도 있다.


실제로 `데이터타입을 어떻게 디자인하느냐` 하는 것은 문제를 해결하는 방법과 밀접한 연관이 있다.

데이터타입을 디자인 하는데 있어서 아주 중요한 개념 하나가 **추상 데이터 타입**이다.
추상 데이터 타입은 Abstract Data Type의 약자로 ADT라고 불린다. 1970년대부터 회자되기 시작하였으며, `데이터 타입을 정의하여 문제를 해결하는 수단으로 사용할 수 있다`는 것에 착안하여 나오기 시작하였다. (ADT의 개념이 먼저 나왔고 그 이후에 OOP가 등장하였다.)



추상 데이터 타입의 정의는 다음과 같다.

An abstract data type (ADT) is a mathematical model for data types. An abstract data type is defined by its behavior (semantics) from the point of view of a user, of the data, specifically in terms of possible **values**, possible **operations** on data of this type, and the behavior of these operations.

추상 데이터 타입은 데이터 타입에 대한 수학적 모델이다. 말을 풀어보자면, 데이터를 사용하는 사용자의 입장에서 사용될 데이터가 무엇이고, 그 데이터가 어떤식으로 동작해야 할지에 대해 **개념적으로 정의하는 것**이다.

이 개념에는 두 가지가 포함되어 있는데 바로 값(values)과 동작(operations)이다.
ADT는 그 데이터 타입에 저장되는 데이터 즉, 데이터로 표현되는 구체적이고 실제적인 **값**(value)은 무엇인지, 그 데이터를 어떻게 만지작, 조작(manipulation)하여 작업할 수 있는지를 의미하는 **동작**(operations)을 개념적으로 정의(명시)하는 것을 의미한다.

- 리스트를 예로 들어보자.
- 리스트로 사용될 수 있는 데이터들은 숫자,문자,리스트,딕셔너리,튜플, 셋 등 다양한 데이터들을 값(values)으로 가질 수 있다.
- 리스트의 동작으로는 데이터 추가, 데이터 삽입, 데이터 삭제, 데이터 합치기, 데이터 정렬 등의 동작을 가질 수 있다.
- 이러한 값과 동작을 개념적으로 정의하면 ADT가 되는 것이고, 이를 실제 구현하면 `리스트`라는 자료구조(Data structure)가 되는 것이다.

## 스택(stack) ADT
- ADT의 가장 대표적이고 기본적인 예로 언급되고 있는 것이 스택이다. 스택의 ADT을 정의해보자.
- 스택은 '쌓아 올린다'는 의미로 데이터를 밑바닥부터 차례대로 쌓아올리는 구조이다.
- https://en.wikipedia.org/wiki/Stack_(abstract_data_type)


### 스택의 특징
- 스택 구조에서 무엇이 필요한지 특징을 먼저 나열해보자.
  - 스택은 위로 쌓아 올리는 것이므로 한 방향으로만 데이터를 넣는다.
  - 데이터를 뺄 때에도 가장 위에 있는 데이터부터 뺀다.
  - 쌓아올린 데이터 중간에 값을 끼워 넣을 수 없다.
  - 스택이 비어 있으면 빼낼 데이터가 없으므로 확인이 필요하다.
- 나중에 들어온 데이터가 가장 빨리 나가므로 후입선출방법(Last In First Out, LIFO)으로 데이터가 처리된다.

### 스택의 값(values)
- 스택에서 사용가능한 값(데이터)으로는 숫자, 문자, 데이터 묶음, 작업 등 무엇이든지 가능하다.
- 스택이 활용될 수 있는 예로는 다음과 같은 작업이 있다.
  - **브라우저에서 뒤로가기**: 웹페이지 방문기록을 순서대로 저장해 둔 다음에 '뒤로가기' 버튼을 누르면 이전 웹페이지를 보여준다.
  - **실행취소**: 실행한 작업을 순서대로 기록해 두었다가 ctrl+z, undo와 같이 실행취소, 되돌리기 버튼을 누르면 마지막에 실행된 것부터 실행을 취소한다.
  - **역순 문자열 만들기**: 맨 끝의 문자열부터 차례대로 만든다.
  - **재귀적 알고리즘**: 함수가 호출된 순서를 기록해 놓았다가 함수에서 탈출할 때 이전 함수로 되돌아간다.

- 위와 같은 작업에서 스택을 활용할 수 있으므로 이러한 작업에서 사용될 가능성이 있는 데이터가 값(values)이 된다.

### 스택의 동작(operations)
- 스택처럼 동작하려면 필요한 기본적인 연산이 무엇이 있는지 정의해보자.
  - `push(item)`: 아이템을 스택에 추가하는 함수(필수)
  - `pop()`: 가장 위에 있는 아이템을 삭제하는 함수. 항상 가장 위에 있는 아이템을 제거하므로 아이템을 명시하지 않아도 된다.(필수)
  - `peek()`: 가장 위에 있는 아이템을 확인하려는 목적으로 아이템을 제거하지는 않고 값을 반환하는 함수
  - `isEmpty()`: 스택이 비었는지 판단하는 함수
  - `size()`: 스택의 아이템의 개수를 반환하는 함수

- 사족을 달자면, 스택구조는 리스트만큼 유연(flexible)하지는 않다. 파이썬에는 리스트라는 엄청 쓰기 좋은 데이터타입이 이미 있으므로 스택이라는 타입을 굳이 만들 필요는 없다. 그럼에도 불구하고 스택을 정의하는 이유는 프랙티컬(practical)한 이유보다는 개념적인 쪽에 훨씬 더 가깝다. 우리가 풀어야 할 여러 가지 문제 중에 이러한 스택이라는 추상화타입(abstract type)을 사용하면 문제를 풀기 편리한 경우가 많이 존재한다. 따라서 사용하기 편하고 불편하고의 문제가 아니라 개념적인 것에 가깝다는 것을 상기하자.


### 스택의 구현
- 개념적으로 스택이라는 ADT를 정의해 봤으니, 실제 구현하여 스택 자료구조로 만들어보자.
- 파이썬뿐만 아니라 대부분의 프로그래밍 언어에서는 이와 같은 ADT를 구현하기 위한 language feature(언어적 특징, 개념, 기능)가 존재한다.
- 파이썬에서는 ADT를 구현하기 위한 language feature가 바로 앞 서 다루었던 클래스(class)와 객체(object)이다.
- 따라서 클래스 Stack을 정의하자.

#### `__init__()`
- 먼저 스택 자료구조에서 초기화하기 위한 ```__init__()``` 메서드를 정의해보자.
- Stack 객체를 만들면 초기화할 변수는
  - 스택의 top 위치
  - 데이터를 저장할 공간
- 여기서 데이터를 저장할 공간으로 리스트를 활용한다.

````python
class Stack:

  def __init__(self):
    '''
    새로운 Stack을 생성한다.
    '''
    self.top = -1
    self.sList = []
````

````python
s = Stack()
print(s)
````

#### `__str__()`
- ```__str__()```를 활용하여 Stack을 출력했을 때 출력되는 형식을 추가해서 지정해보자.
- 데이터 저장할 공간을 리스트로 만들고 있기 때문에 출력하면 리스트 형태로 출력된다. [],{},() 모두 리스트, 딕셔너리, 튜플로 각각 사용되고 있는 기호이므로 우리는 S< >S로 만들어보자.
- Stack 클래스에 아래의 `__str__()` 메서드를 추가하자.

````python
def __str__(self):
  '''
  Stack의 출력 형식을 S<item1, item2, item3, ...>S과 같이 출력한다.
  '''
  value=str(self.sList)[1:-1]
  return "S<"+value+">S"
````

#### `push(item)`
- push(item) 메서드를 정의해보자.
- self.sList 리스트에 item을 추가한 후 아이템의 상단 위치인 top을 1 증가 시킨다.

````python
def push(self, item):
  '''
  Stack에 item을 넣는다.
  '''
  self.sList.append(item)
  self.top += 1
````

#### `isEmpty()`
- Stack에 isEmpty() 메서드를 정의해 보자. Stack에 item이 없으면 True를, 있으면 False를 반환한다.

````python
def isEmpty(self):
  '''
  Stack에 item이 없으면 True, 있으면 False를 반환한다.
  '''
  if self.top == -1:
    return True
  return False
````

#### `pop()`
- Stack에 item이 있으면 item을 반환하고 Stack에서는 제거한다. 이는 리스트의 pop() 메서드가 동일한 기능을 제공하므로 이를 활용한다.
- 만약 item이 없으면 'Stack is Empty'를 반환한다.

````python
def pop(self):
  '''
  Stack에 item이 있으면 item을 반환하고 Stack에서는 제거한다.
  Stack이 비어 있으면 'Stack is Empty'를 반환한다.
  '''
  if not(self.isEmpty()):
    self.top -= 1
    return self.sList.pop()
  return 'Stack is Empty'
````

#### `peek()`
- peek()은 맨 위에 아이템을 확인하기 위해 주로 사용한다. 맨 위에 아이템을 반환한다.(삭제는 하지 않는다.)

````python
def peek(self):
  '''
  맨 위의 아이템을 반환한다.
  '''

  if not(self.isEmpty()):
    return self.sList[-1]
````

#### `size()`
- Stack의 아이템의 개수를 반환한다.

````python
def size(self):
  return len(self.sList)
````

### 스택의 활용
- 구현한 스택자료구조를 활용하여 괄호 매칭 문제를 해결하는 수단으로 사용해 보자.
- 함수나 연산식을 작성할 때 혹은 리스트나 딕셔너리와 같은 데이터 타입을 정의할 때도 대괄호, 중괄호, 소괄호가 사용된다. 이때 사용되는 괄호는 쌍이 맞아야 한다.
- 괄호의 쌍이 맞다는 의미는
  - 여는 괄호의 종류, 개수와  닫는 괄호의 종류, 개수가 같아야 한다.
  - 가장 늦게 열린 괄호가 가장 먼저 닫혀야 한다.
  - 괄호가 열리기 전에 닫히면 안된다.
- 스택을 활용하여 괄호의 쌍이 올바른지 판단하는 함수 bracket()를 작성해보자.

print(bracket('([)]')) #괄호의 순서가 잘못되어 Invalid
print(bracket('([])')) #Valid
print(bracket('][')) #괄호가 열리기 전에 닫혀 Invalid

## 큐(Queue) ADT
- 큐는 먼저 들어온 데이터가 먼저 나가는 구조이다. 선입선출(FIFO: First in First out)방법으로 데이터가 다루어진다.


### 큐의 특징
- 큐는 후단(rear)으로만 데이터가 들어오고, 전단(front)으로만 데이터가 나간다.

  - 중간의 데이터를 끼워 넣거나 뺄 수 없다.
  - 큐가 비어 있으면 빼낼 데이터가 없으므로 확인이 필요하다.
  

<div align="center"><img src="https://haesunbyun.github.io/common/images/adt6.png" width= 400 style="width:400px;"></div>

### 큐의 값(values)
- 큐의 접근방법을 사용하는 데이터(숫자, 문자, 데이터 묶음, 작업) 들은 무엇이든지 가능하다.

- 큐가 활용될 수 있는 예로는 다음과 같은 작업이 있다.
  - **매표소나 은행창구의 대기열**: 먼저 온 사람이 먼저 서비스를 받는다.
  - **통신에서의 데이터 패킷들의 모델링**: 트래픽 조절과 혼잡 제어를 위한 큐잉 이론에 기저한 분석 모델링
  - **프린터와 컴퓨터 사이의 버퍼링**: 컴퓨터에서 인쇄로 보낸 문서 대기열 순서대로 서비스를 받는다.

- 위와 같은 작업에서 큐를 활용할 수 있으므로 이러한 작업에서 사용될 가능성이 있는 데이터가 값(values)이 된다.

### 큐의 동작(operations)
- 큐처럼 동작하려면 필요한 기본적인 연산이 무엇이 있는지 정의해보자.
  - `__init__()`: 큐를 초기화한다.
  - `enqueue(e)`: 주어진 아이템을 큐의 맨 뒤에 추가한다.
  - `dequeue()`: 큐가 비어있지 않으면 맨 앞 아이템을 삭제하고 반환한다.
  - `isEmpty()`: 큐가 비어있으면 True를 아니면 False를 반환한다.
  - `peek()`: 가장 앞에 있는 아이템을 확인하려는 목적으로 큐가 비어있지 않으면 맨 앞 요소를 삭제하지 않고 반환한다.
  - `size()`: 큐의 아이템 개수를 반환한다.

### 큐의 구현

````python
class Queue:

  def __init__(self):
    self.qList = []

  def __str__(self):
    value=str(self.qList)[1:-1]
    return f"Q<{value}>Q"

  def isEmpty(self):
    return not bool(self.qList)

  def enqueue(self, item):
    self.qList.append(item)

  def dequeue(self):
    if not(self.isEmpty()):
      return self.qList.pop(0)

  def peek(self):
    if not(self.isEmpty()):
      return self.qList[0]

  def size(self):
      return len(self.qList)
````

````python
q = Queue()
q.isEmpty()
q.enqueue('hi')
q.enqueue('hello')
q.enqueue('python')
print(q)
q.dequeue()
print(q)
print(q.peek())
q.size()
````

## Simple 3차원 벡터(vector) ADT
- 벡터는 수학, 물리학, 공학에서 크기와 방향을 함께 갖는 물리량을 의미한다.
- 프로그래밍에서 벡터는 한 가지 타입의 여러 개 원소를 변수에 저장하고, 원소의 개수에 따라 크기가 동적으로 변하는 배열을 의미한다.
- 벡터화 연산은 벡터의 동일 인덱스에 위치한 원소들끼리 연산을 수행하는 기능이다.
- 벡터는 크기가 동적으로 변하는 배열인 리스트의 특징을 가지고 있지만, 리스트는 벡터연산을 지원하지 않는다.

````python
a=[1,2,3,4]
b=[5,6,7,8]
print(a+b)
````

- 파이썬에서는 Numpy 모듈에서 벡터연산을 지원한다. 따라서 벡터라는 데이터 타입 또한 굳이 만들 필요는 없다.
- 그렇지만 개념적으로 3차원 벡터의 합, 벡터의 곱을 구하는 ADT를 정의하고 구현해보자.

````python
import numpy as np
a = np.arange(20,50,10) # 20,30, 40
b = np.arange(1,4)

print(a)
print(b)
print('벡터의 합:', a+b)
print('벡터의 곱:',a.dot(b)) #20*1 + 30*2 + 40*3
````

### 벡터의 동작
벡터의 합과 벡터의 곱 연산만 정의해보자. 이때 예약메서드 가운데 덧셈 예약메서드 `__add__()`와 곱셈 예약메서드 `__mul__()` 메서드를 재정의 하여 사용해보자.



$v_1 = (20,30,40)$

$v_2 = (1,2,3)$

$v_3 = v_1 + v_2$

$var = v_1 * v_2$

- `__add__()`: 동일 인덱스에 위치한 원소들끼리 덧셈 연산을 수행한다.
- vector 타입으로 반환한다.

````python
class vector3:
  def __init__(self, x, y, z):
    self.x = x
    self.y = y
    self.z = z

  def __add__(self, v):
    return vector3(self.x + v.x, self.y + v.y, self.z + v.z)

v1 = vector3( 20, 30, 40 )
v2 = vector3( 1, 2, 3 )
v3 = v1+v2
print(v3)
````

- print(v3)를 했을 때 바로 결과가 나오도록 `__str__()`를 정의해보자.

````python
class vector3:
  def __init__(self, x, y, z):
    self.x = x
    self.y = y
    self.z = z

  def __add__(self, v):
    return vector3(self.x + v.x, self.y + v.y, self.z + v.z)

  def __str__(self):
    return f'{self.x} {self.y} {self.z}'

v1 = vector3( 20, 30, 40 )
v2 = vector3( 1, 2, 3 )
v3 = v1+v2
print(v3)
````

- `__mul__()`: 벡터 곱셈 연산을 수행한다.

````python
class vector3:
  def __init__(self, x, y, z):
    self.x = x
    self.y = y
    self.z = z

  def __add__(self, v):
    return vector3(self.x + v.x, self.y + v.y, self.z + v.z)

  def __str__(self):
    return f'{self.x} {self.y} {self.z}'

  def __mul__(self, v):
    return (self.x*v.x) + (self.y*v.y) + (self.z*v.z)

v1 = vector3( 20, 30, 40 )
v2 = vector3( 1, 2, 3 )
v3 = v1+v2
print(v3)
var = v1*v2
print(var)
````

그 밖의 메서드 이름은 아래의 문서를 참고하기 바란다. 아래 문서에 특수(special method)라는 이름으로 설명되어 있다.
특수 매서드, 매직 매서드, 예약 매서드 등으로 불린다.

https://docs.python.org/ko/3.7/reference/datamodel.html#special-method-names

````python
class vector3:
  def __init__(self, x, y, z):
    self.x = x
    self.y = y
    self.z = z

  def __str__(self):
    return f'{self.x} {self.y} {self.z}'

  def add(self, v):
    self.x += v.x
    self.y += v.y
    self.z += v.z

  def dotProduct(self, v):
    return (self.x*v.x) + (self.y*v.y) + (self.z*v.z)

v1 = vector3( 20, 30, 40 )
v2 = vector3( 1, 2, 3 )
print(v2.dotProduct(v1))
````
