## 그래프 용어

- **정점(Vertex 또는 Node)**: 데이터를 저장하는 위치
- **간선(Edge)**: 정점를 연결하는 선. 링크(Link) 또는 브랜치(branch) 로도 불린다.

```
  - 하나의 간선은 두 개의 정점을 연결한다.
  - 그래프 G는 G=(V,E)로 표현, V는 정점의 집합이며, E는 간선의 집합이다.
  - 정점 A와 B를 연결하는 간선을 (A,B)로 표현한다.
  - 정점 A와 B로 간선의 방향이 있는 경우 <A,B>로 표현한다.
```

- **방향 그래프(Directed Graph)**: 간선에 방향이 있는 그래프

- **차수(Degree)**: 정점에 인접한 정점의 수
  - 방향 그래프에서는 정점으로 들어오는 차수를 진입 차수(In-degree), 정점에서 나가는 차수를 진출 차수(Out-degree)로 구분한다.
  - 방향 그래프는 진입 차수와 진출 차수가 다를 수 있다. <$v_i, v_j$> $\neq$  <$v_j, v_i$>
  - 무방향 그래프는 <$v_i, v_j$> = <$v_j, v_i$>이다.

- **경로(Path)**: 시작 정점 $u$부터 도착 정점 $v$까지의 정점들을 나열한 표현한다.
  - 단순경로: 경로 상의 정점들이 모두 다른 경로
  - [A,B,D,C]: 시작 정점 A로부터 도착 정점 C까지의 단순 경로
  
- **싸이클(Cycle)**: 시작 정점과 도착 정점이 동일한 단순 경로
  - [A,B,D,C,A]

- **완전 그래프(Complete graph)**: 그래프에서 간선의 수가 최대인 그래프
  - 무방향 그래프의 경우 $n$개의 정점이 있다면 간선의 수는 $n(n-1)/2$
    - (예)정점이 4개인 그래프에서 간선의 최대 개수는 6

  - 방향 그래프의 경우 $n$개의 정점이 있다면 간선의 수는  $n(n-1)$

  <div align="center"><img src="https://haesunbyun.github.io/common/images/graph4.PNG" width=600></div>

- **인접(adjacent)**: 무방향 그래프에서 정점 a,b에 대하여 간선(a,b)가 있으면, 정점 a는 정점 b에 인접한다고 한다.
- **부속(incident)**: 무방향 그래프에서 정점 a,b에 대하여 간선(a,b)가 있으면, 간선(a,b)는 정점 a와 b에 부속한다고 한다.

- **가중치(Weighted) 그래프**: 간선에 가중치가 부여된 그래프
  - 가중치는 두 정점 사이의 거리, 시간 등이 될 수 있음


- 트리(Tree): 싸이클이 없는 그래프
- 신장트리(Spanning Tree): 그래프의 모든 정점들을 싸이클없이 연결하는 부분그래프

## 그래프 탐색 알고리즘
- 그래프 탐색은 하나의 정점에서 시작하여 모든 정점을 한 번씩 방문하는 작업이다.
  - 예, 전자회로 연결 판단, 미로 출구 찾기

- 대표적인 그래프 탐색 알고리즘은 **깊이 우선 탐색(DFS, depth first search)**와 **너비 우선 탐색(BFS, breath first search)** 알고리즘이 있다.




### 깊이 우선 탐색(DFS, depth first search)
  - 그래프에서 깊이 탐색을 우선적으로 탐색하는 알고리즘
  - 동작과정
    1. 임의의 정점에서 시작하여 이웃하는 하나의 정점을 방문한다. 이때 방문한 정점은 반드시 방문되었다고 표시를 해야 한다.
    2. 방금 방문한 정점의 이웃 정점을 방문하며, 이웃하는 정점을 모두 방문한 경우에는 이전 정점으로 되돌아 간다.
    3. 거기서 다시 아직 방문하지 않은 인접 정점을 찾아 동일한 방법의 탐색을 수행하는 방식이다.

  - 스택이나 재귀함수를 이용하여 구현 가능하다.
    1. 탐색 시작 노드를 스택에 삽입하고 방문 처리
    2. 스택의 최상단 노드에 방문하지 않은 인접한 노드가 하나라도 있으면 그 노드를 스택에 넣고 방문 처리.방문하지 않은 인접 노드가 없으면 스택에서 최상단 노드를 꺼냄
    3. 더 이상 2번의 과정을 수행할 수 없을 때까지 반복

#### DFS 알고리즘 재귀적 구현

````python
def dfs_recursive(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    print(start, end=' ')

    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs_recursive(graph, neighbor, visited)

# 그래프 표현 (인접 리스트)
# input으로 입력하기 귀찮아서 미리 넣어놓음
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A', 'D', 'E'],
    'D': ['B', 'C', 'F'],
    'E': ['C', 'G', 'H'],
    'F': ['D'],
    'G': ['E', 'H'],
    'H': ['E', 'G']
}

# DFS 호출
dfs_recursive(graph, next(iter(graph)))
````

### 너비 우선 탐색(BFS, breath first search)
- 그래프에서 가까운 노드부터 우선적으로 탐색하는 알고리즘
- 시작 정점으로부터 가까운 정점을 먼저 방문하고 멀리 떨어져 있는 정점을 나중에 방문하는 순회 방법이다.
- 가까운 거리에 있는 정점들을 차례대로 저장하고, 들어간 순서대로 꺼낼 수 있는 자료구조인 큐를 이용하여 구현이 가능하다.

### 스택을 이용한 DFS 크롤링 예제

````python
from requests import request
from requests.compat import urlparse, urljoin
from bs4 import BeautifulSoup

# 시작 URL 설정
seed = 'https://haesunbyun.github.io/Core-Computing-Ver.2/intro.html'
base = urlparse(seed)[1]  # 기본 도메인 추출
tovisit = [seed]  # 방문할 URL 리스트 초기화, 스택
visited = []  # 방문한 URL 리스트 초기화

while tovisit:
    url = tovisit.pop()  # 방문할 URL 리스트에서 마지막 URL을 가져옴
    page = request('get', url)  # 해당 URL에 GET 요청을 보냄
    soup = BeautifulSoup(page.content, 'lxml')  # 응답 내용을 BeautifulSoup을 사용해 파싱
    visited.append(url)  # 현재 URL을 방문한 URL 리스트에 추가

    for _ in soup.select('a[href]'):  # 페이지 내의 모든 링크를 순회
        new_url = urljoin(seed, _['href'])  # 절대 URL로 변환
        # 링크가 페이지 내부 링크이거나 같은 도메인의 링크가 아닌 경우 건너뜀
        if _['href'][0] == '#' or urlparse(new_url)[1] != base:
            continue
        # 새로운 URL이 방문 예정 또는 방문한 URL 리스트에 없는 경우에만 추가
        if new_url not in visited + tovisit + [seed+'/']:
            tovisit.append(new_url) # 새 URL을 방문할 URL 리스트에 추가

# 방문한 URL 리스트 출력
visited
````

### 스택을 이용한 DFS 시 탐색범위 제한
- 뉴스 사이트처럼 문서들이 관련 주제에
따라 계층형 관계로 구성되어 있을 때, 특정 링크를 따라 계속해서 문서들을 추적하다 보면 결국 원래의 주제에서 멀리 떨어진 내용들을 다루게 된다.
- 따라서 탐색 범위 제한을 통해 무한히 확장하는 인터넷 공간상에서 사이클에 빠지지 않도록 **비 순환성(Acyclic)**으로 만들거나, 시드로부터 너무 멀리 떨어져 있어서 발생할 수 있는 내용적 **비 유사성**을 사전에 방지하고 크롤러가 적당한 시점에 탈출 할 수 있도록 참조 범위를 기준으로 제한할 필요가 있다.
- 아래의 코드는 스택을 이용한 깊이 우선 탐색 시, 시드를 0부터 시작하여 하이퍼링크를 통해 다음 페이지를 탐색할 때마다 순차적으로 1씩 증가하였고, 최대 참조
범위가 2 미만일 때 까지 만으로 제한한 예이다.

````python
from requests import request
from requests.compat import urlparse, urljoin
from bs4 import BeautifulSoup

# 시작 URL 설정
seed = 'https://news.naver.com/'
base = urlparse(seed)[1]  # 기본 도메인 추출
tovisit = [(seed,0)]  # 방문할 URL 리스트 초기화
visited = []  # 방문한 URL 리스트 초기화

while tovisit:
    url = tovisit.pop(-1)  # 방문할 URL 리스트에서 첫 번째 URL을 가져옴
    page = request('get', url[0])  # 해당 URL에 GET 요청을 보냄
    soup = BeautifulSoup(page.content, 'lxml')  # 응답 내용을 BeautifulSoup을 사용해 lxml문서로 파싱
    visited.append(url)  # 현재 URL을 방문한 URL 리스트에 추가

    for _ in soup.select('a[href]'):  # 페이지 내의 모든 링크를 순회
        new_url = urljoin(seed, _['href'])  # 절대 URL로 변환

        # 링크가 비어있거나, 링크가 페이지 내부 링크이거나 같은 도메인의 링크가 아닌 경우 건너뜀
        if not _['href'] or _['href'][0] == '#' or new_url[:4] != 'http' or url[1] > 1:
          continue
        # 새로운 URL을 방문 예정 또는 방문한 URL 리스트에 없는 경우에만 추가
        if new_url not in [_[0] for _ in visited] + [_[0] for _ in tovisit] + [seed + '/']:
          tovisit.append((new_url, url[1]+1))

    # 오래 걸려서 10개까지만 제한
    if len(visited)>=10:
      break

# 방문한 URL 리스트 출력
visited
````

### 큐를 이용한 BFS 시 탐색범위 제한
- 탐색한 도메인을 지정함으로서 탐색 범위를 제한한 예이다.

````python
from requests import request
from requests.compat import urlparse, urljoin
from bs4 import BeautifulSoup

# 시작 URL 설정
seed = 'https://haesunbyun.github.io/Core-Computing-Ver.2/intro.html'
seed = 'https://news.naver.com/'
base = urlparse(seed)[1]  # 기본 도메인 추출
tovisit = [seed]  # 방문할 URL 리스트 초기화
visited = []  # 방문한 URL 리스트 초기화
enabled= ['_sources'] # 탐색을 허용하는 도메인 서브

while tovisit:
    url = tovisit.pop(0)  # 방문할 URL 리스트에서 첫 번째 URL을 가져옴
    page = request('get', url)  # 해당 URL에 GET 요청을 보냄
    soup = BeautifulSoup(page.content, 'lxml')  # 응답 내용을 BeautifulSoup을 사용해 lxml문서로 파싱
    visited.append(url)  # 현재 URL을 방문한 URL 리스트에 추가

    for _ in soup.select('a[href]'):  # 페이지 내의 모든 링크를 순회
        new_url = urljoin(seed, _['href'])  # 절대 URL로 변환

        # 링크가 비어있거나, 링크가 페이지 내부 링크이거나 같은 도메인의 링크가 아닌 경우 건너뜀
        if _['href'][0] == '#':
          continue
        if True not in map(lambda x:x in enabled, urlparse(new_url)[2].split('/')):
          continue
        # 새로운 URL을 방문 예정 또는 방문한 URL 리스트에 없는 경우에만 추가
        if new_url not in visited + tovisit + [seed + '/']:
          tovisit.append(new_url)

# 방문한 URL 리스트 출력
visited
````
