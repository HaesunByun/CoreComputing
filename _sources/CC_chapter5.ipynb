## 클래스와 객체





### 객체란?
- **객체**는 데이터와 그 데이터에 관련되는 동작(절차, 방법, 기능)을 모두 포함한 개념이다.
- 이 세상에 존재하는 모든 것, 예를 들어, 학생, 자동차, 책상, 버튼 등 모든 것이 객체가 될 수 있으며, 모든 객체는 속성(attribute)과 동작(action, method)을 가지고 있다.
- 예로, 자동차라는 객체는 메이커, 모델, 색상, 연식, 가격 등의 속성을 가지고 있고, 주행하기, 방향바꾸기, 주차하기, 음성인식 등의 동작을 한다.

### 클래스란?
- 파이썬에서는 `클래스를 통해 객체를 구현`한다.
- 클래스는 서로 관련된 데이터(속성, 값)와 해당 데이터를 다루는 함수(동작)들을 하나로 묶어 추상화한 것으로, 객체를 정의하기 위한 템플릿(폼, 포맷, 양식), 객체에 대한 설계도라고 볼 수 있다.

### 클래스 정의하기

- 클래스를 정의할 때에는 `class`라는 키워드로 시작한다.
```
class 클래스명:
     클래스 변수

     인스턴스 메서드(method)
```
- `클래스 변수`는 동일 클래스의 모든 인스턴스가 공통(공유)으로 값을 가지는 변수를 말한다. 일반적으로 인스턴스 메서드(함수)의 바깥쪽에 정의한다.

- `인스턴스 메서드`는 클래스의 함수를 의미한다. 첫 번째 인자는 항상 self를 가진다.


- 실 세계의 객체인 개(Dog)를 생각해 보도록 하자.

- 저 마다의 개들은 견종이 있고, 색깔이 있고, 꼬리가 있고, 다리가 4개이고, ...등 속성이 있다.
- '개'라는 객체가 공통적으로 가지는 동작(operations)은 꼬리를 흔들고, 네 다리로 뛰며, 짖기도 한다.
- 우리는 '개'라는 객체를 개념적으로 위와 같이 **추상화**를 할 수 있다.


- 그럼 '개'라는 객체를 소프트웨어적으로 실체를 만들기 위해 클래스를 통해 템플릿을 먼저 만들어보자.

#### 클래스 변수 정의
- 클래스안에 변수 attr1과 attr2를 만들어보자.  attr1과 attr2는 클래스 변수이다. 클래스 변수는 동일 클래스에서 만들어지는 모든 인스턴스가 공통적으로 갖는 값이다.
- Dog 클래스를 통해 만들어지는 모든 인스턴스는 'mammal'(포유류)과 'dog'(개)라는 값(value)을 갖는다.

````python
class Dog:
    attr1 = "mammal"
    attr2 = "dog"
````

#### 인스턴스 메서드 정의
- 클래스 Dog에 **인스턴스 메서드**를 만들어보자.
- 인스턴스 메서드는 클래스의 함수이므로 함수를 정의하는 것과 동일하게 만든다.
```
def show(self):
    인스턴스 메서드 본문
```
여기서 show()메서드는 첫 번째 인자로 self를 갖는다. self는 인스턴스 그 자신을 의미한다. 동일 클래스의 인스턴스를 여러 개 만들 수 있으므로 각 인스턴스 자신을 가르키는 self가 필요하다. 아직 self가 이해되지 않더라도 실망하지 말자. 대학교 2학년 때 쯤 C++로 객체지향프로그래밍을 처음으로 배우기 시작했을 때 self를 이해하는데 한참 걸렸던것 같다.ㅜㅜ
- 메서드안에서는 self.클래스변수명으로 클래스 변수의 값에 접근한다.
- 클래스를 정의했으면 파이썬이 인식하도록 아래 셀을 실행시켜준다.

````python
class Dog:
    '''
    Dog Class by haesun
    '''
    attr1 = "mammal"
    attr2 = "dog"

    def show(self):
        print("I'm a", self.attr1)
        print("I'm a", self.attr2)
````

### 인스턴스 생성과 활용
- 아주 심플하지만 클래스를 정의해 놓았으니 이제 그 클래스의 인스턴스를 만들어보자.
- 인스턴스를 만드는 방법도 간단하다.

#### 인스턴스 생성
```
인스턴스명 = 클래스명()
```
- 지금까지 dictVar = dict() 또는 listVar = list()와 같이 딕셔너리 타입의 변수, 리스트 타입의 변수를 만들어 사용했었다. 이것은 사실 dict() 클래스의 인스턴스, list() 클래스의 인스턴스를 만들었던 것이다.
- 예를 들어, Dog() 클래스의 인스턴스 momo를 만드려면 momo = Dog()와 같이 쓴다. 이를 통해 Dog라는 객체가 인스턴스화 된다.


````python
momo = Dog()       # an object is instantiated
````

#### 클래스 변수 접근
- 클래스안에 정의된 클래스 변수에 접근할 때에는 다음과 같이 점(.)으로 연결한다.
```
인스턴스명.클래스 변수명
```

- 예를 들어, Dog 객체는 attr1, attr2 두 개의 클래스 변수를 갖고 있는데 이 가운데 attr1의 값에 접근하려면 momo.attr1와 같이 쓴다.



````python
print(momo.attr1)
print(momo.attr2)
````

#### 인스턴스 메서드 호출
- 클래스안에 정의된 인스턴스 메서드를 호출하려면 다음과 같이 쓴다.
```
인스턴스명.메서드명()
클래스명.메서드명(인스턴스명)
```
- 예를 들어, Dog 객체는 show()라는 인스턴스 메서드를 갖고 있는데 show() 메서드를 호출하려면 momo.show()와 같이 쓴다.
- 이는 또한 여러분들이 listVar = list()를 만들어 놓고 listVar.append(item) 했던 것과 같은 원리이다.
- 인스턴스명이 아니라 클래스명으로도 직접 호출할 수 있다. 예를 들어, Dog.show(momo)와 같이 호출할 수도 있다. 이 때에는 show()괄호안에 인스턴스명을 명시해줘야 한다.


````python
momo.show()
Dog.show(momo)
````

#### 인스턴스 변수
- 인스턴스 변수는 각 인스턴스마다 값을 가지는 변수를 말한다.
- 메서드에서 정의되며 `self.변수명`으로 접근한다.
- 예를 들어, 아래 인스턴스 메서드 run()에서 self.speed는 인스턴스 변수이다.

````python
class Dog:
    '''
    Dog Class by haesun
    '''
    attr1 = "mammal"
    attr2 = "dog"

    def show(self):
        print("I'm a", self.attr1)
        print("I'm a", self.attr2)
        print(f"My speed is {self.speed}")

    def run(self, speed):
        self.speed = speed
        self.show() #class안에서 호출
````

- 인스턴스 변수는 동일 클래스 내 정의된 메서드 어디에서든지 `self.변수명`으로 활용 가능하다.
- 예를 들어,  self.speed 변수는 run() 메서드에서 만들어졌지만 show() 메서드에서 self.speed 값에 접근할 수 있다.
- 그래서 지역변수라고 하지 않고 인스턴스 변수라고 한다.

````python
momo = Dog()
momo.run(10)
momo.show()
````

- 인스턴스 변수를 클래스 밖에서 접근하려면 다음과 같이 쓴다.
```
인스턴스명.인스턴스변수
```

````python
momo.speed = 50
momo.speed
````

### 예약 메서드와 속성 활용하기


#### 속성과 메서드 확인
- 객체와 관련된 속성과 메서드를 확인하려면 dir() 함수를 사용한다.
- 예를 들어, dir(Dog)를 통해 Dog 객체가 가지고 있는 속성과 메서드를 확인해보자. 정의하지 않은 무언가가 많이 나와서 당황스러웠을 것이다.
- 파이썬에서 클래스를 정의하면 object라는 최상위 클래스를 상속받는다. 상속받는다는 것은 object 클래스가 갖고 있는 속성과 메서드를 정의하는 클래스에서도 다 갖게 된다는 의미이다.
- 상속받은 이름에는 앞뒤로 이중언더바(__)가 붙는다. 이들 가운데 어느 것이 속성(변수)이고 어느 것이 메서드(함수)인지는 눈으로 구분되지는 않는다.
- 가장 마지막 쯤에는 Dog 클래스에서 직접 정의한 attr1, attr2, run, show도 들어가 있는 것도 확인된다.

````python
print(dir(Dog))
````

- 한 예로 `__class__`라는 속성을 확인해보자.
- `__class__`는 객체의 데이터 타입을 보여준다.
- 변수 a를 만들고 a의 타입을 확인해 보니 <class 'int'>로 `int` 타입이다.
- 변수 momo의 타입을 확인해보니 <class '__main__.Dog'>로 `__main__.Dog` 타입이다. 여기서 `__main__`은 파이썬의 최상위 모듈이다. 최상위 모듈은 파이썬 프로그램의 실행 시작점으로 최상위 모듈안에 Dog 클래스가 정의되었다는 의미이다.

````python
a = 100
print(a.__class__)
print(momo.__class__)
````

- 참고로, 변수뿐만 아니라 객체를 만들기위해서 사용되는 클래스 자체도 데이터 타입이 있다.
- 타입을 확인해 보니, int 클래스나 Dog 클래스 모두 <class 'type'>로 type 클래스에 속하는 인스턴스임을 알 수 있다.

````python
print(int.__class__)
print(Dog.__class__)
````

#### 예약 메서드
- 상속받은 것 들 중에는 메서드도 많다. 이 가운데 이중언더바(__)가 붙어있는 메서드들을 예약 메서드(reserved method) 또는 매직 메서드(magic method)라고 한다. 즉 이미 특정 기능으로 예약되어 이름 붙여진 메서드이다.
- 이와 같은 예약 메서드들을 여러분들이 만든 클래스에서 동일 이름으로 재정의하면 클래스와 관련된 특정 기능을 수행 시킬 수 있다. 이 가운데 몇 개의 예약 메서드를 활용해보자.
- `__init__` : 생성자라고도 불리우며, 인스턴스가 생성 될때 자동으로 실행된다.
- `__del__` : 소멸자라고도 불리우며, 인스턴스 삭제 시 자동으로 실행된다.
- `__call__` : 인스턴스명( )으로 호출했을 때 실행된다.
- `__str__` : 객체 자체를 출력할 때 출력 형식을 지정한다.



#### `__init__` 메서드
- 생성자 메서드인 `__init__` 메서드를 재정의해보자.

#### `__str__` 메서드

#### `__call__`메서드
- `인스턴스명()`와 같이 함수처럼 사용할 때 호출된다.

````python
class Dog:
    '''
    Dog Class by haesun
    '''
    attr1 = "mammal"
    attr2 = "dog"

    def __init__(self, breed, color, sound):
        self.breed = breed  #견종
        self.color = color  #색깔
        self.sound = sound  #소리
        self.speed = 0

    def __str__(self):
        return f"🐕 I'm a {self.breed}, {self.color}, {self.sound}, {self.speed}🐕"

    def __call__(self, x, y):
        print(f"🐕 멍멍 왈왈 나 불렀어요? 저는 똑똑한 강아지에요. 🐕" )
        self.x = x
        self.y = y
        return print(f'답은 {self.x + self.y}입니다.')

    def show(self):
        print("I'm a", self.attr1)
        print("I'm a", self.attr2)
        print(f"I'm a {self.breed}")
        print(f"My color is {self.color}")
        print(f"My sound is {self.sound}")
        print(f"My speed is {self.speed}")

    def run(self, speed):
        self.speed = speed
````

````python
momo = Dog("poodle", "brown","bowwow")
momo(5, 3)
````

#### `__del__`메서드
- 객체가 소멸될때 호출된다.

````python
class Dog:
    '''
    Dog Class
    '''
    attr1 = "mammal"
    attr2 = "dog"

    def __init__(self, breed, color, sound):
        self.breed = breed  #견종
        self.color = color  #색깔
        self.sound = sound  #소리
        self.speed = 0

    def __str__(self):
        return f"🐕 I'm a {self.breed}, {self.color}, {self.sound}, {self.speed}🐕"

    def __call__(self, x, y):
        print(f"🐕 멍멍 왈왈 나 불렀어요? 저는 똑똑한 강아지에요. 🐕" )
        self.x = x
        self.y = y
        return print(f'답은 {self.x + self.y}입니다.')

    def __del__(self):
        print(f"🐕안녕히 계세요. 여러분~~~~ 저는 행복을 찾아 떠납니다!! 🐕" )

    def show(self):
        print("I'm a", self.attr1)
        print("I'm a", self.attr2)
        print(f"I'm a {self.breed}")
        print(f"My color is {self.color}")
        print(f"My sound is {self.sound}")
        print(f"My speed is {self.speed}")

    def run(self, speed):
        self.speed = speed
````

````python
momo = Dog("poodle", "brown","bowwow")
del momo
````

## 클래스와 객체지향프로그래밍(OOP)

클래스는 객체지향프로그래밍에서 유래되었다. 객체지향프로그래밍은 1980년대부터 회자되기 시작한 프로그래밍 패러다임중에 하나이다.

- **프로그래밍 패러다임**에는 크게 다음과 같이 세 가지로 나눌 수 있다.
  - 절차적(Procedural, imperative) 프로그래밍
  - 객체지향(Object Oriented) 프로그래밍
  - 함수형(Functional) 프로그래밍




### 절차적 프로그래밍
이전까지 여러분들이 했던 프로그래밍은 절차적 프로그래밍이었다.

예를 들어 자판기를 절차적 프로그래밍으로 코딩한다고 생각해보자.  
- 자판기는 루프를 돌면서 입력을 기다리고 있을 것이다.
- 동전이 들어오면 액수만큼 잔액에 더한다.
- 그리고 금액을 화면에 표시한다.
- 사용자가 어떤 버튼을 누르냐에 따라서 if ~ elif ~ elif ~ elif ~ else를 따라 다르게 동작한다.



이것은 자판기가 동작하는 방식을 절차적으로 바라보는 입장이다.
자판기가 동작하는 방식의 흐름대로, 즉 플로우가 흘러가는 방법을 플로우차트로 그리라고 했다면 자판기의 동작이 간단할 때는 그럭저럭 그릴 수 있다.


그러나 자판기의 동작이 조금만 복잡해지면 프로세스 또한 복잡해진다. 복잡한 프로세스를 따라간다는게 쉽지 않으며 인간이 잘하지 못하는 영역이기도 하다. 오류를 발견하기도 쉽지 않다.

### 객체지향 프로그래밍

그렇다면 프로세스를 절차적으로 바라보지 않고 달리 바라볼 수 있는 방법이 있을까? 그래서 나온 것이 객체지향 프로그래밍의 패러다임이다. 객체지향 프로그래밍의 패러다임은 `세상을 컴퓨터적으로 바라보는 패러다임이다.`


객체지향 프로그래밍(OOP)에서는 모든 대상이 객체(Object)가 된다.
각 객체는 이벤트가 생겼을 때 자신이 어떤 일을 해야 하는지만 알고 있고 전체의 프로세스가 어떻게 되는지는 알 필요가 없다. 예를 들어, 자판기에서의 버튼은 객체이고 각 객체는 자신이 해야 할 역할이 명백하게 정해져 있다.


객체지향 프로그래밍의 예를 가장 잘 보여주는 것이 윈도우의 GUI(graphic user interface)이다. 윈도우가 대표적인 OOP의 수혜자이며 안드로이드든, IOS든 모두 GUI기반으로 동작하고 있다. 마이크로소프트(MS)나 애플(Apple)이 상업적인 목적으로 GUI를 도입하여 OOP 컨셉을 들고 나오면서 GUI가 현실적으로 개발되는데 성공했고 그러면서 OOP와 GUI는 뗄레야 뗄 수가 없는 관계가 되었다.
GUI는 버튼을 누르면 메뉴가 펼쳐지거나 버튼을 누르면 동작을 하거나 수많은 버튼이 있지만 각 버튼들은 자기가 맡은 역할 밖에 모른다.


### 함수형 프로그래밍
- 함수형 프로그래밍은 함수가 코드 구성의 기본이며 모든 것을 함수로 사용하겠다는 것이다.
- 함수파트에서 다루었던 고차함수도 함수형 프로그래밍의 한 형태이다.
  - 함수의 전달인자로 함수를 준다거나 (예, map(int, ['1','2','3']),
  - 함수를 반환값으로 준다거나, return mySum()
  - 함수를 변수에 할당한다거나 등
  ```
  a = print
  a('와우! 이게 된다고?')               
  ```          
- 코드를 함수로만 작성하면 가독성을 높이고 유지보수를 용이하게 해준다.
- 함수형 프로그래밍은 이 밖에도 많은 특징이 있지만 여기서 다루지 않는다.

````python
a = print
a('와우! 이게 된다고?')
````

파이썬 프로그래밍 언어는 절차적 프로그래밍, 객체지향 프로그래밍, 함수적 프로그래밍 모두 지원한다. 요즘 나오는 프로그래밍 언어는 대부분 이 세가지를 다 지원한다.
그렇다면 객체지향프로그래밍은 클래스만들고 메서드만들면 다 되는 것인가? 아니다. 그것보다 고급(advanced) 기능들이 더 많이 포함된다.
오늘 다뤘던 클래스는 객체지향 프로그래밍을 가능하게 하는 하나의 도구일 뿐이다.
객체지향프로그래밍의 추가적인 고급 기능들은 여기서 다루는 범위를 벗어나므로 추후에 기회가 있을 때 다루기로 한다.
