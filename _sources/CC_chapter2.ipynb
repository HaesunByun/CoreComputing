### Deepcopy()
- deepcopy() í•¨ìˆ˜ëŠ” copy ëª¨ë“ˆì— ìˆë‹¤.
- Deep copyëŠ” ë³µì‚¬ë³€ìˆ˜ì˜ ê°’ì„ ì‘ì—…(ìˆ˜ì •, ì‚­ì œ, ì¶”ê°€ ë“±)í•´ë„ ì›ë³¸ë³€ìˆ˜ì˜ ê°’ì´ ì‘ì—…ë˜ì§€ ì•ŠëŠ”ë‹¤.

````python
a=[1,2,3,[10,20,30]]
b={'a':1, 'b':2, 'c':[10,20,30]}
c=np.array([{'a':[1,2]},{'b':[1,2]},{'c':[1,2]}])
````

````python
import copy
a1=copy.deepcopy(a)
b1=copy.deepcopy(b)
c1=copy.deepcopy(c)
a1[3][0] = 100
b1['c'][0] = 100
c1[0]['a'][0] = 100

print('Copy Data ìˆ˜ì • í›„ origin data=============')
print(a)
print(b)
print(c)

print('\nCopy Data ìˆ˜ì • í›„ copy data=============')
print(a1)
print(b1)
print(c1)
````
