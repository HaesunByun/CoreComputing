### ë²„ë¸”ì •ë ¬ êµ¬í˜„
- ë²„ë¸”ì •ë ¬ ì•Œê³ ë¦¬ì¦˜ì„ êµ¬í˜„í•´ë³´ì.
- ì •ë ¬ì„ ìœ„í•œ ë°ì´í„° 20ê°œë¥¼ 0ë¶€í„° 100 ì‚¬ì´ì˜ ì„ì˜ì˜ ìˆ«ìë¡œ ë§Œë“¤ì–´ ë³€ìˆ˜ dataì— ì €ì¥í•˜ì.

````python
import random

def randomData20():
  data=[]
  for i in range(20):
    data.append( random.randint(0,100) )
  return data

data=randomData20()
print(data)
````

````python
def mybubbleSort(data):

  length = len(data)-1
  # ì „ì²´ stepì€ n-1ë²ˆ ë°˜ë³µ, nì€ ë°ì´í„° ê°¯ìˆ˜
  for i in range(length):
    for j in range(length-i): #í•œ stepë§ˆë‹¤ ë¹„êµ íšŸìˆ˜ê°€ 1ì”© ê°ì†Œ
      if data[j] > data[j+1]: #ì¸ì ‘í•œ ë‘ ì›ì†Œë¥¼ ë¹„êµ
        data[j], data[j+1] = data[j+1], data[j] #ìë¦¬ë°”ê¿ˆ

data=randomData20()
print(f'Before: {data}')
mybubbleSort(data)
print(f'After: {data}')
````

````python
def mySelectionSort(data):
    length = len(data)-1
    for i in range(length): # ì „ì²´ stepì€ n-1ë²ˆ ë°˜ë³µ
      idxMin=i              # ì´ˆê¸° ìµœì†Ÿê°’
      for j in range(i+1,length+1): # í•œ stepë§ˆë‹¤ ë¹„êµ íšŸìˆ˜ê°€ 1ì”© ê°ì†Œ
        if data[idxMin] > data[j]:
          idxMin = j                # ìµœì†Œê°’ì´ë©´ ê·¸ë•Œì˜ ì¸ë±ìŠ¤ë¥¼ ì €ì¥

      data[i],data[idxMin]=data[idxMin], data[i] #ìë¦¬ë°”ê¿ˆ

data=randomData20()
print(f'Before: {data}')
mySelectionSort(data)
print(f'After: {data}')
````

````python
def myinsertionSort(data):
  #TODO
  length = len(data)
  for i in range(1, length): # ì „ì²´ stepì€ n-1ë²ˆ ë°˜ë³µ
    j = i - 1                # ì •ë ¬ëœ ë²”ìœ„ì˜ ë ì¸ë±ìŠ¤
    key = data[i]            # ì •ë ¬ë˜ì§€ ì•Šì€ ë²”ìœ„ì˜ ì‹œì‘ ë°ì´í„°
    # (ì •ë ¬ëœ ë²”ìœ„ì˜ ë ë°ì´í„°ê°€ keyë³´ë‹¤ í¬ê³ ) and
    # (ì •ë ¬ëœ ë²”ìœ„ì˜ ë ì¸ë±ìŠ¤ê°€ 0ë³´ë‹¤ í¬ê±°ë‚˜ ê°™ìœ¼ë©´)
    while (data[j] > key) and (j >= 0):
      data[j+1] = data[j] # data[j]ë¥¼ data[j+1]ìë¦¬ì— ì €ì¥. ì¦‰, ë’¤ë¡œ ë°€ì–´ ì €ì¥
      j = j - 1
    data[j+1] = key        # keyë¥¼ ì‚½ì…

data=randomData20()
print(f'Before: {data}')
myinsertionSort(data)
print(f'After: {data}')
````
