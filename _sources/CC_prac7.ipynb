<font color='blue'>`Test Case-1`</font>

Output Sample
```

```

- 파이썬의 isnumeric( )이라는 메서드는 자연수형태의 문자열인 경우 True를, 실수, 문자열, 음수형태의 문자열인 경우 False를 반환합니다. 예를 들어, '10'.isnumeric()은 True, '10.0'.isnumeric(), 'hi'.isnumeric(), '-10.0'.isnumeric()  
은 False를 반환합니다. 이 메서드를 활용하셔도 좋습니다.
