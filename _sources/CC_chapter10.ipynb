## 트리 용어

- 아래는 트리 구조를 보여주고 있다.
<div align="center"><img src="https://haesunbyun.github.io/common/images/tree2.PNG" width=400></div>

- 트리(Tree): 계층적으로 표현된 데이터 구조이다.
- 노드(Node): 노드는 트리의 구성 요소이다.(A,B,C,D,E,F,G,H,I,J)
- 루트 노드(Root node): 트리의 최상위 노드로, 부모 노드가 없다.(A)
  - 루트 노드를 기준으로 여러 개의 자식 노드(child node)를 가질 수 있다.
  - 각 노드는 0개 이상의 자식 노드를 가질 수 있으며, 각 자식 노드는 다시 하위 자식을 가질 수 있다.
  - 루트 노드에서부터 시작하여 각 노드가 계층적으로 연결되어 있다.
- 리프 노드(Leaf node) 또는 단말(terminal node): 자식 노드가 없는 노드. 즉, 트리의 끝을 나타냄(E,F,G,H,I,J)
- 비단말 노드: 자식을 가지는 노드(A,B,C,D)
- 서브트리(Subtree): 특정 노드를 루트로 하는 트리 구조, 트리의 일부
- 차수(Degree): 어떤 노드가 가지고 있는 자식의 수
- 깊이(Depth): 루트 노드에서 특정 노드까지의 경로 길이
- 높이(Height): 특정 노드에서 가장 깊은 리프 노드까지의 경로 길이





## 트리의 표현

- 컴퓨터에서 트리를 표현하는 가장 일반적인 방법은 아래와 같이 데이터 필드와 링크 필드로 표현하는 것이다.
- 데이터 필드는 항목 값을 저장하며, 링크 필드는 자식 노드를 가르킨다.

<div align="center"><img src="https://haesunbyun.github.io/common/images/tree3.PNG" width=300></div>

- 그런데 일반적인 트리에서 각 노드들은 임의의 개수의 자식 노드를 가질 수 있으며 이는 링크필드의 길이가 달라져야 한다.

- 이러한 단점을 해결하는 한가지 방법은 연결 리스트를 사용하여 트리를 표현할 수도 있다. 각 노드가 첫 번째 자식노드를 가르키는 링크 필드와 다음 형제노드를 가리키는 링크 필드를 갖고 있는 방법이다.

## 이진트리

- 연결 리스트를 사용하여 트리를 표현하는 방법은 매우 복잡하다. 다행히 실제로는 두 개의 자식 노드를 사용하는 이진트리(Binary tree)가 많이 사용된다. 이 방법은 구현하기가 편리하다.
<div align="center"><img src="https://haesunbyun.github.io/common/images/tree4.PNG" width=200></div>

- 이진 트리(Binary tree)
  - 모든 노드가 2개의 서브 트리를 갖는 트리
  - 모든 노드의 차수가 2 이하로 최대 2개까지의 자식 노드를 가질 수 있다.
  - 서브 트리는 공집합 일 수 있다.
  - 즉, 이진 트리의 정의는
    - (1) 공집합이거나
    - (2) 루트와 왼쪽 서브 트리, 오른쪽 서브 트리로 구성된  노드들의 유한 집합

- 포화 이진 트리(Full binary tree)
  - 모든 노드가 0개 또는 2개의 자식 노드를 가지는 이진 트리
  - 모든 리프 노드는 동일한 깊이에 있음
  - 최대 노드 수: 높이 $h$인 포화 이진 트리의 총 노드 수는 $2^h -1$개

<div align="center"><img src="https://haesunbyun.github.io/common/images/tree5.PNG" width=300></div>

- 완전 이진 트리(Complete binary tree)
  - 모든 레벨이 채워져 있지만 마지막 레벨은 왼쪽에서 오른쪽으로 순차적으로 채워진 트리
  - 포화 이진 트리는 항상 완전 이진 트리이지만, 완전 이진 트리는 반드시 포화 이진 트리는 아니다.

## 이진 트리의 순회
- 트리를 **순회(traversal)**한다는 것은 트리에 속하는 모든 노드를 한 번씩 방문하여 데이터를 어떤 목적에 맞게 처리한다는 것을 의미한다.
- 트리와 같은 비선형구조는 다양한 순회방법이 존재한다.
- 트리 순회 방법은 루트(V)와 왼쪽 서브트리(L), 오른쪽 서브 트리(R)를 어떤 순서로 방문하느냐에 따라 전위, 중위, 후위 방법이 있다.
- 예를 들어, 다음과 같은 트리가 있다고 가정한다.
  <div><img src="https://haesunbyun.github.io/common/images/tree9-0.PNG" width=300></div>
- 아래 트리에서는 노드 위에 트리를 방문하는 순서를 번호로 나타내고 있다.

    

### 전위 순회(Preorder traversal): VLR
    - 루트를 먼저 방문하고 그 다음에 왼쪽 서브트리를 방문하고 오른쪽 서브트리를 마지막으로 방문함
  <div><img src="https://haesunbyun.github.io/common/images/tree9.PNG" width=300></div>
    - 예: 구조화된 문서 출력
  <div><img src="https://haesunbyun.github.io/common/images/tree16.PNG" width=400></div>
   

### 중위 순회(Inorder traversal): LVR
    - 왼쪽 서브트리, 루트, 오른쪽 서브트리 순으로 방문한다.
  <div><img src="https://haesunbyun.github.io/common/images/tree10.PNG" width=300></div>



### 후위 순회(Postorder traversal): LRV
    - 왼쪽 서브 트리, 오른쪽 서브트리, 루트 순으로 방문한다.

## 이진 트리 순회 코드

### 전위 순회

- TreeNode라는 클래스로 노드를 표현한다.
  - self.value에는 data를 저장한다.
  - self.left는 왼쪽 자식노드를 가르킨다.
  - self.right는 오른쪽 자식노드를 가르킨다.
<div align="center"><img src="https://haesunbyun.github.io/common/images/tree4.PNG" width=200></div>

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
```

```python
def preorder_traversal(node):
    if node is None:
        return []

    result = []
    # 현재 노드 방문
    result.append(node.value)
    # 왼쪽 서브트리 방문
    result.extend(preorder_traversal(node.left))
    # 오른쪽 서브트리 방문
    result.extend(preorder_traversal(node.right))

    return result

# 예제 이진 트리 생성
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

# 전위 순회 수행: VLR
preorder_result = preorder_traversal(root)
print(preorder_result)  # 출력: [1, 2, 4, 5, 3, 6, 7]
```

### 중위 순회

```python
# 중위 순회 수행: LVR
def inorder_traversal(node):
    if node is None:
        return []

    result = []
    # 왼쪽 서브트리 방문
    result.extend(inorder_traversal(node.left))
    # 현재 노드 방문
    result.append(node.value)
    # 오른쪽 서브트리 방문
    result.extend(inorder_traversal(node.right))

    return result

inorder_result = inorder_traversal(root)
print(inorder_result)
```

### 후위 순회

```python
# 후위 순회 수행: LRV
def postorder_traversal(node):
    if node is None:
        return []

    result = []
    # 왼쪽 서브트리 방문
    result.extend(postorder_traversal(node.left))
    # 오른쪽 서브트리 방문
    result.extend(postorder_traversal(node.right))
    # 현재 노드 방문
    result.append(node.value)

    return result

postorder_result = postorder_traversal(root)
print(postorder_result)  # 출력: [4, 5, 2, 6, 7, 3, 1]
```

## 이진 트리 응용: 수식 트리
- 산술식이나 논리식을 트리형태로 표현한 것을 수식 트리라고 한다.
- 수식 트리에서는 피연산자들이 리프(단말)노드가 되고 연산자는 비단말 노드가 된다.
- 아래는 산술식과 논리식을 수식 트리로 표현한 예이다.
  - $a+b$
<div><img src="https://haesunbyun.github.io/common/images/tree12.PNG" width=300></div>
  - $a-(b*c)$
<div><img src="https://haesunbyun.github.io/common/images/tree13.PNG" width=300></div>
  - $(a<b)$ or $(c>d)$
<div><img src="https://haesunbyun.github.io/common/images/tree14.PNG" width=300></div>  








- 이러한 수식 트리를 전위 순회, 중위 순회, 후위 순회의  방법으로 읽으면 각각 전위 표기식, 중위 표기식, 후위 표기식이 된다.
- 일반적으로 수식은 중위표기식으로 표현한다. 그러나 중위 표기식에서는 연산자 우선순위를 고려해야 하고, 괄호를 사용하여 이를 명확히 해야한다.

  예: $a+b$, $a-(b*c)$, $(a<b)$ or $(c>d)$

- 수식트리를 후위 순회 방법으로 읽으면 후위 표기식이 된다.

  예: $ab+$, $abc*-$, $ab<cd>or$

- 후위 표기식은 연산자 우선순위를 명시할 필요가 없고, 괄호를 사용하지 않아 수식의 해석을 단순하게 만들어준다. 또한 연산이 나타나는 순서대로 계산을 진행하면 되므로 스택을 통해 구현이 쉽다.

```python
# 수식 트리 생성과 순회
root = TreeNode('+')
root.left = TreeNode('a')
root.right = TreeNode('b')

preorder_result = preorder_traversal(root)
print('전위순회:', preorder_result)

inorder_result = inorder_traversal(root)
print('중위순회:', inorder_result)

postorder_result = postorder_traversal(root)
print('후위순회:', postorder_result)
```

```python
# 수식 트리 생성과 순회
root = TreeNode('-')
root.left = TreeNode('a')
root.right = TreeNode('*')
root.right.left = TreeNode('b')
root.right.right = TreeNode('c')

preorder_result = preorder_traversal(root)
print('전위순회:', preorder_result)

inorder_result = inorder_traversal(root)
print('중위순회:', inorder_result)

postorder_result = postorder_traversal(root)
print('후위순회:', postorder_result)
```

## 이진탐색트리(BST, Binary Search Tree)

- 우리는 앞서 이진 탐색 알고리즘을 학습했다. 이진 탐색은 정렬된 배열에서 수행되고, 주로 정렬된 정적 데이터에서 사용된다.
- 이번에 학습할 이진 탐색 트리는 트리 구조로 이루어져 있으며 동적으로 데이터가 변경되는 경우에 적합하다.
- 이진탐색트리
  - 키(key): 데이타를 구별하기 위한 고유한 값
  - 모든 노드는 유일한 키를 갖는다.
  - 왼쪽 서브 트리 키들은 루트 키보다 작다.
  - 오른쪽 서브 트리의 키들은 루트의 키보다 크다.
  - 왼쪽과 오른쪽 서브 트리도 이진탐색트리이다.

<div><img src="https://haesunbyun.github.io/common/images/tree19.PNG" width=300></div>  

    - 왼쪽 서브 트리에 있는 값들 3,7,12는 루트 노트인 23보다 작다.
    - 오른쪽 서브 트리에 있는 값들 26, 31, 27은 루트 노드인 23보다 크다.
    - 이 성질은 모든 노드에서 만족된다.
    - 이 트리를 중위 순회 방법으로 순회하면 3,7,12,23,26,27,31로 오름차순으로 노드를 방문한다.

### 탐색연산

```python
class TreeNode:
    def __init__(self, key):
        self.value = key
        self.left = None
        self.right = None
```

- target_value가 루트의 value(키)와 같으면 그 루트가 찾는 노드이다.
- target_value가 루트의 value보다 작으면 찾는 노드는 왼쪽 서브트리에 있다. 왼쪽을 기준으로 다시 탐색을 시작한다.
- target_value가 루트의 value보다 크면 찾는 노드는 오른쪽 서브트리에 있다. 오른쪽을 기준으로 다시 탐색을 시작한다.

```python
def search(root, target_value):
    if root is None or root.value == target_value:
        return root
    if target_value < root.value:
        return search(root.left, target_value)
    if target_value > root.value :
        return search(root.right, target_value)
```

### 삽입연산
- 노드의 삽입을 위해 삽입할 위치를 먼저 탐색해야 한다.
- 탐색에 실패한 위치가 삽입할 위치가 된다.
  - 9를 삽입
<div><img src="https://haesunbyun.github.io/common/images/tree20.PNG" width=300></div>



```python
def insert(root, value):
    if root is None:
        return TreeNode(value)
    else:
        if root.value < value:
            root.right = insert(root.right, value)
        else:
            root.left = insert(root.left, value)
    return root
```

```python
# 예제
root = TreeNode(23)
print(root.value)
root = insert(root, 7)
root = insert(root, 26)
root = insert(root, 3)
root = insert(root, 12)
root = insert(root, 31)
root = insert(root, 27)

# 검색
result = search(root, 27)
if result:
    print("노드 찾음:", result.value)
else:
    print("노드 없음")
```

## 이진탐색트리의 성능 분석
- 이진탐색트리는 트리의 높이를 $h$라고 했을 때 $O(h)$가 된다.
- $n$개의 노드를 가진 이진탐색트리의 경우, 높이가 $log_2n$이므로, 이진탐색트리의 평균 시간 복잡도는 $O(log_2n)$이다.
- 예를들어, 백만개의 데이터의 경우 $log_2(1000000) ≈ 20$번만에 결과를 찾는다.

## 우선순위 큐
- 소방차나 119 구급차와 같은 긴급 차량들은 다른 차들에 비해 우선순위(priority)를 갖는다. 컴퓨터에서도 우선순위의 개념이 필요할 때가 있다.
- 우선순위 큐는 우선순위의 개념을 큐에 도입한 자료구조로 **우선순위가 높은 자료를 먼저 꺼내는 구조이다**.
- 이러한 우선순위 큐를 구현하는 가장 효율적인 방법은 힙(heap)이다.
- **힙(heap)**은 완전 이진 트리(Complete Binary Tree)의 한 종류로, 최댓값이나 최솟값을 빠르게 찾아야 하는 알고리즘에서 유용하게 쓰인다.
- 아래 트리는 최대 힙 트리와 최소 힙 트리를 보여준다.
  - 최대 힙 트리: 부모 노드의 값이 항상 자식 노드의 값보다 크거나 같은 구조를 가진 완전 이진 트리

  - 최소 힙 트리: 부모 노드의 값이 항상 자식 노드의 값보다 작거나 같은 완전 이진 트리

- 파이썬에서는 heapq 모듈을 사용하여 우선순위 큐를 구현할 수 있다.
- 아래 코드는 우선순위 큐에 우선순위와 항목을 삽입하고, 삭제하는 예이다.
- heapq 모듈은 최소 힙을 기반으로 가장 작은 요소가 루트에 위치하며, 각 연산에서 리스트를 힙처럼 사용할 수 있다.

```python
import heapq

# 리스트를 heap처럼 사용
priority_queue = []

# 힙에 삽입(heap, (우선순위, 항목))
heapq.heappush(priority_queue, (2, "task 2"))
heapq.heappush(priority_queue, (1, "task 1"))
heapq.heappush(priority_queue, (3, "task 3"))

# 가장 우선순위가 높은 (작은) 요소를 꺼냄
print(heapq.heappop(priority_queue))  # 출력: (1, "task 1")
```

## 마무리

- 트리는 계층적인 자료를 표현하는데 이용되는 자료 구조이다.
- 트리는 두 개의 자식 노드를 사용하는 이진트리(Binary tree)가 많이 사용된다. 이진 트리(Binary tree)는 모든 노드가 2개의 서브 트리를 갖는 트리이다.이진트리는 연결 리스트를 사용하여 구현하기가 편리하다.
- 이진 트리의 순회는 전위, 중위, 후위가 있다.
- 이진 탐색 트리는  동적으로 데이터가 변경되는 경우에 적합하다.
