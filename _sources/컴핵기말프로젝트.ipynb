# 기말 프로젝트 안내
- 단순한 코딩 기술만을 요구하는 시대는 지나갔으며, 현실적인 문제 해결에 필요한 창의성, 협업 및 문제 해석 능력을 키우는 것이 중요해졌습니다.
- 기존의 폐쇄적인 시험 방식을 벗어나, 오픈북 형태로 학생들이 다양한 자료를 참조하고, 서로의 코드를 이해하며 개선할 수 있는 능력을 발전시키기 위해 기말고사 방식을 변경하였습니다.
- 이 프로젝트에서는 다른 사람의 코드를 분석하고 이해하며, 필요한 경우 수정하여 더 나은 솔루션을 제안할 수 있는 창의적 사고를 키우는 것이 목표입니다.
- 이 과정을 통해 학생들은 현실 세계의 문제에 접근하는 방식을 학습하고, 이를 통해 더욱 효과적으로 문제를 해결할 수 있는 능력을 키울 수 있습니다.

## 가. 프로젝트 목표와 기대효과

### 1.프로젝트 목표

- **알고리즘 이해 증진**: 다양한 알고리즘과 자료구조를 이해하고, 실제 문제에 적용할 수 있도록 함

- **코드 리팩토링 및 최적화 능력 향상**: 제출된 코드를 분석하고 개선하는 과정을 통해, 코드의 효율성과 가독성을 높일 수 있는 능력 개발

- **창의적 문제 해결 능력 개발**: 기존 문제를 변형하거나 확장하여 새로운 문제를 생성함으로써, 창의적 사고와 문제 해결 능력을 키움

- **피드백 능력 강화**: 피드백을 제공하고 받는 과정을 통해 상호 작용 능력 증진

- **학술적 정직성의 실천**: 백준, LeetCode 등의 원본 문제를 사용할 때, 데이터를 활용할 때, 정확한 출처를 명시하고, 저작권을 존중하는 행동 강화


### 2.프로젝트 기대효과

- **실용적인 코딩 기술 향상**: 실제 문제를 해결하고, 코드를 개선하는 경험을 통해 학생들의 코딩 능력 향상

- **창의적 문제 해결 능력의 향상**: 다양한 문제에 접근하고, 기존의 해결책을 재해석하며 새로운 솔루션을 창출할 수 있는 능력 개발

- **코드 분석 및 개선 능력 증진**: 다른 학생의 코드를 효과적으로 분석하고 이해하며, 필요한 개선을 도출할 수 있는 실력 향상

- **학술적 정직성 인식 증진**: 정확한 출처 표기와 저작권 존중의 중요성 인식

- **자기 주도적 학습 능력 증진**: 스스로 문제를 해결하고, 새로운 문제를 창출하는 과정을 통해 자기 주도적 학습 능력 증진
