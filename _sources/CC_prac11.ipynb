- 이진 트리의 순회는 전위, 중위, 후위 방법이 있다.
- 전위 순회(Preorder traversal): VLR
    - 루트를 먼저 방문하고 그 다음에 왼쪽 서브트리를 방문하고 오른쪽 서브트리를 마지막으로 방문함
  <div><img src="https://haesunbyun.github.io/common/images/tree9.PNG" width=300></div>

    

- 중위 순회(Inorder traversal): LVR
    - 왼쪽 서브트리, 루트, 오른쪽 서브트리 순으로 방문한다.
  <div><img src="https://haesunbyun.github.io/common/images/tree10.PNG" width=300></div>



- 후위 순회(Postorder traversal): LRV
    - 왼쪽 서브 트리, 오른쪽 서브트리, 루트 순으로 방문한다.

## 이진 트리 응용: 수식 트리
- 산술식이나 논리식을 트리형태로 표현한 것을 수식 트리라고 한다.
- 수식 트리에서는 피연산자들이 리프(단말)노드가 되고 연산자는 비단말 노드가 된다.
- 아래는 산술식과 논리식을 수식 트리로 표현한 예이다.
  - $a+b$
<div><img src="https://haesunbyun.github.io/common/images/tree12.PNG" width=300></div>
  - $a-(b*c)$
<div><img src="https://haesunbyun.github.io/common/images/tree13.PNG" width=300></div>
  - $(a<b)$ or $(c>d)$
<div><img src="https://haesunbyun.github.io/common/images/tree14.PNG" width=300></div>  








- 이러한 수식 트리를 전위 순회, 중위 순회, 후위 순회의  방법으로 읽으면 각각 전위 표기식, 중위 표기식, 후위 표기식이 된다.
- 일반적으로 수식은 중위표기식으로 표현한다. 그러나 중위 표기식에서는 연산자 우선순위를 고려해야 하고, 괄호를 사용하여 이를 명확히 해야한다.

  예: $a+b$, $a-(b*c)$, $(a<b)$ or $(c>d)$

- 수식트리를 후위 순회 방법으로 방문하면 후위 표기식이 된다.

  예: $ab+$, $abc*-$, $ab<cd>or$

- 후위 표기식은 연산자 우선순위를 명시할 필요가 없고, 괄호를 사용하지 않아 수식의 해석을 단순하게 만들어준다. 또한 연산이 나타나는 순서대로 계산을 진행하면 되므로 스택을 통해 구현이 쉽다.

### 3-1. 만들어진 트리정보 출력하기
* 3-1 ~ 3-3은 트리를 시각화 해보기 위한 참고코드입니다. 우리 수업의 주요 코드는 아닙니다.

만든 트리의 모양을 확인해보기 위해 nodes와 edges 리스트를 만들어본다.


````python
def tree_to_lists(root):
    nodes = []
    edges = []

    def traverse(node):
        if node is None:
            return
        # 노드 값을 nodes 리스트에 추가
        nodes.append(node.value)

        # 왼쪽 자식이 있으면 edges에 추가하고 재귀적으로 왼쪽 자식 탐색
        if node.left:
            edges.append((node.value, node.left.value))
            traverse(node.left)

        # 오른쪽 자식이 있으면 edges에 추가하고 재귀적으로 오른쪽 자식 탐색
        if node.right:
            edges.append((node.value, node.right.value))
            traverse(node.right)

    traverse(root)
    return nodes, edges
````

````python
nodes, edges = tree_to_lists(tree)

print("Nodes:", nodes)
print("Edges:", edges)
````

### 3-2. 그래프로 그리기 위해 포지션 설정
- 위에서 만든 노트와 엣지들을 이진 트리 모양으로 그리기 위해 적절한 포지션을 정한다.

````python
def compute_positions(root):
    pos = {}

    # 트리의 위치를 재귀적으로 설정하는 함수
    def assign_positions(node, x=0, y=0, layer=1):
        if node is None:
            return

        # 현재 노드의 위치를 pos 딕셔너리에 추가
        pos[node.value] = (x, y)

        # 왼쪽 자식 노드 위치 설정
        if node.left:
            assign_positions(node.left, x - 1 / layer, y - 1, layer + 1)

        # 오른쪽 자식 노드 위치 설정
        if node.right:
            assign_positions(node.right, x + 1 / layer, y - 1, layer + 1)

    # 루트 노드부터 위치 설정 시작
    assign_positions(root)
    return pos
````

````python
pos = compute_positions(tree)
print("Positions:", pos)
````
