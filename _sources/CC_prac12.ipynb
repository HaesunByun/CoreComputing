## Problem: 친구 관계 찾기 (BFS)

- 이번에는 BFS를 활용하여 친구를 찾고, 친밀도를 파악해보자. 직접 친한 경우 친밀도는 1이고,
한 명 건너 친구 관계라면 1씩 더해진다.

- 친구 관계 정보는 아래와 같다.

1. Summer와 John은 친구 관계 입니다.
2. Summer와 Justin은 친구 관계 입니다.
3. Summer와 Mike는 친구 관계 입니다.
4. Justin과 May는 친구 관계 입니다.
5. May와 Kim은 친구 관계 입니다.
6. John과 Justin은 친구 관계 입니다.
7. Justin과 Mike는 친구 관계 입니다.
8. Tom과 Jerry는 친구 관계 입니다.

- 친구 관계를 그래프로 표현하면 아래와 같다.
- 중복된 관계를 고려하지 않기 위해 양방향 그래프를 구성한다.
- 양방향 그래프로 친구 관계 그래프를 표현하면 중복 계산을 최소화하고 최단 경로를 찾음
<div><img src="https://haesunbyun.github.io/common/images/graph_friends2.PNG" width=500></div>

```
def build_graph(graph_dict):
    # 양방향 그래프로 변환
    graph = defaultdict(set)
    for person, friends in graph_dict.items():
        for friend in friends:
            graph[person].add(friend)
            graph[friend].add(person)
    return graph
```
