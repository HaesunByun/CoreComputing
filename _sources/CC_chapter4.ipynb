## Goodì˜ˆ) íŒ©í† ë¦¬ì–¼(Factorial)
- ì¬ê·€í•¨ìˆ˜ì˜ ê°€ì¥ ëŒ€í‘œì ì¸ ì˜ˆë¡œ í™œìš©ë˜ëŠ” íŒ©í† ë¦¬ì–¼ì„ ì •ì˜í•´ë³´ì.
- íŒ©í† ë¦¬ì–¼ì€ *n*ì´ í•˜ë‚˜ì˜ ìì—°ìˆ˜ì¼ ë•Œ, 1ì—ì„œ *n*ê¹Œì§€ì˜ ëª¨ë“  ìì—°ìˆ˜ì˜ ê³±ì„ *n*ì— ìƒëŒ€í•˜ì—¬ ì´ë¥´ëŠ” ë§ë¡œ *n!*ë¡œ ì½ëŠ”ë‹¤.
- ì˜ˆë¥¼ ë“¤ì–´, *5!*ì„ ë°˜ë³µì ì¸ í˜•íƒœ(iterative form)ìœ¼ë¡œ ì“°ë©´ $ 5!= 5*4*3*2*1 $ ì´ë‹¤.
- ì´ë¥¼ ìˆ˜ì‹ìœ¼ë¡œ ë‚˜íƒ€ë‚´ë©´ ì•„ë˜ì™€ ê°™ë‹¤.



- Iterative form

$n! = n\times (n-1)\times \cdots \times 1$




- ì½”ë“œë¡œ ì‚´í´ë³´ì.
- ì•„ë˜ì˜ factorial_iterative(n) í•¨ìˆ˜ëŠ” ë°˜ë³µë¬¸ì„ ì‚¬ìš©í•˜ì—¬ íŒ©í† ë¦¬ì–¼ì„ êµ¬í•˜ëŠ” ì½”ë“œì´ë‹¤.

````python
def factorial_iterative(n):
    f=1
    for i in range(n, 0, -1):
        f = f * i
    return f
````

````python
factorial_iterative(5)
````

- ê·¸ëŸ¬ë‚˜ ìˆ˜í•™êµê³¼ì„œì—ëŠ” ì•„ë˜ì™€ ê°™ì´ ì¬ê·€ì ì¸ í˜•íƒœ(recursive form)ë¡œ ì •ì˜ë˜ì–´ ìˆë‹¤.
- ì¬ê·€ì ì¸ í˜•íƒœë¡œ í‘œí˜„í•˜ëŠ” ê²ƒì„ ì„ í˜¸í•˜ëŠ” ì´ìœ ëŠ” ë¬¸ì œê°€ ê°„ë‹¨í•  ë•Œì—ëŠ” ë°˜ë³µì ì¸ í˜•íƒœë¡œ ì •ì˜í•´ë„ ë˜ì§€ë§Œ, ë¬¸ì œê°€ ì¡°ê¸ˆë§Œ ë³µì¡í•´ì§€ë©´ ë°˜ë³µì ì¸ í˜•íƒœë¡œ ì •ì˜í•˜ê¸°ì— ì–´ë ¤ìš´ ìˆœê°„ì´ ì˜¤ê¸° ë•Œë¬¸ì´ë‹¤.
- ì¬ê·€ì ì¸ í˜•íƒœëŠ” ì•„ë˜ì™€ ê°™ì´ ê°„ë‹¨í•˜ê²Œ ì •ì˜í•  ìˆ˜ ìˆë‹¤.



- Recursive form

$n! = n \times (n-1)!~~~$ if $n>1$.

$n!=1~~~$ if $n=1$.

- ì•„ë˜ì˜ factorial_recursive(n)ì€ ì¬ê·€í•¨ìˆ˜ë¥¼ ì´ìš©í•˜ì—¬ íŒ©í† ë¦¬ì–¼ì„ êµ¬í•˜ëŠ” ì½”ë“œì´ë‹¤.
- *n!*ì„ êµ¬í•˜ëŠ” ì•Œê³ ë¦¬ì¦˜ì´ë‚˜ *(n-1)!*ë¥¼ êµ¬í•˜ëŠ” ì•Œê³ ë¦¬ì¦˜ì´ ë™ì¼í•˜ë‹¤.
- ë”°ë¼ì„œ ì¬ê·€ì ìœ¼ë¡œ êµ¬í•  ìˆ˜ ìˆë‹¤.

````python
def factorial_recursive(n):
    if n==1:
        return 1
    else:
        return n * factorial_recursive(n-1)
````

````python
factorial_recursive(5)
````

- sum() í•¨ìˆ˜ì™€ ë™ì¼í•˜ê²Œ ë™ì‘í•˜ëŠ” í•¨ìˆ˜ë¥¼ iterativeí•˜ê²Œ ì •ì˜í•´ë³´ë©´ ì•„ë˜ sum_iterative(data) í•¨ìˆ˜ì™€ ê°™ì´ ì‘ì„±í•  ìˆ˜ ìˆë‹¤.
- ì—¬ê¸°ì„œ ì „ë‹¬ì¸ìëŠ” ë¦¬ìŠ¤íŠ¸ì´ë¯€ë¡œ ë§¤ê°œë³€ìˆ˜ dataëŠ” ë¦¬ìŠ¤íŠ¸ë¥¼ ë°›ëŠ”ë‹¤.

````python
def sum_iterative(data):
    s = 0
    for i in data:
        s += i
    return s
````

````python
sum_iterative([5,7,6,9,2])
````

- sum() í•¨ìˆ˜ì™€ ë™ì¼í•˜ê²Œ ë™ì‘í•˜ëŠ” í•¨ìˆ˜ë¥¼ recursiveí•˜ê²Œ ì •ì˜í•´ë³´ì.
- sum_recursive(data[1:])ì„ í†µí•´ ë¦¬ìŠ¤íŠ¸ [5,7,6,9,2]ëŠ”
  - sum_recursive([7,6,9,2])
  - sum_recursive([6,9,2])
  - sum_recursive([9,2])
  - sum_recursive([2])
  - sum_recursive([ ]) ì™€ ê°™ì€ ìˆœì„œë¡œ ì „ë‹¬ëœë‹¤.
- ê·¸ë¦¬ê³  return ë˜ëŠ” ê°’ì€
  - sum_recursive([ ])ì—ì„œ => 0
  - sum_recursive([2])ì—ì„œ 2+0ì¸ => 2
  - sum_recursive([9,2])ì—ì„œ 9+2ì¸ => 11
  - sum_recursive([6,9,2])ì—ì„œ 6+11 => 17
  - sum_recursive([7,6,9,2])ì—ì„œ 7+17 => 24
  - sum_recursive([5,7,6,9,2])ì—ì„œ 5+24 => 29ê°€ ë°˜í™˜ëœë‹¤.

````python
def sum_recursive(data):
    if len(data)==0:
        return 0
    else:
        return data[0] + sum_recursive(data[1:])
````

````python
sum_recursive([5,7,6,9,2])
````

- ì•„ë˜ fibonacci_iterative(n)í•¨ìˆ˜ì—ì„œ `curr, next = 0, 1` ëŠ” curr = 0, next = 1ì„ ì €ì¥í•œë‹¤ëŠ” ì˜ë¯¸ì´ë‹¤.
- ê·¸ëŸ° í›„ì—, for ë°˜ë³µë¬¸ì—ì„œ `curr, next = next, curr + next`ë¥¼ í†µí•´ curr = next, next = curr+nextë¥¼ ë°˜ë³µí•œë‹¤. ì¦‰ ì´ì „ next ê°’ì´ í˜„ì¬ currê°€ ë˜ê³ , ì´ì „ curr + ì´ì „ next ê°’ì„ ìƒˆë¡œìš´ nextê°’ì—  ë„£ì–´ì¤€ë‹¤.

````python
def fibonacci_iterative(n):
  curr, next = 0, 1
  for _ in range(n):
    curr, next = next, curr + next
  return curr

fibonacci_iterative(5)
````
