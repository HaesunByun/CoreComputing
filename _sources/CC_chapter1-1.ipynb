# Chapter 1-1. 문자열 포맷팅

문자열을 만들때 상용의 문구(글의 구절)는 정해져있고 문구 사이에 예를 들어, 이름, 나이, 주소 등을 바꿔 끼워서 사용하는 일이 빈번하게 발생한다. 문자열 포맷팅은 틀(양식, 포맷)은 갖추어져있고 그 안에 키워드들만 바꿔가면서 작업을 할 수 있도록 하는 것을 의미한다.


## "Old Style" String formatting

변수를 문자열과 함께 출력할 때 연결(concatenate)하는 방법으로 % 기호를 쓴다. 이때 % 뒤에 어떤 알파벳을 붙이느냐에 따라 데이터 타입의 의미가 달라진다. %s, %c, %d, %f 등 %와 함께 쓰는 알파벳을 **형식지정자**라 한다.
- \%s : 문자열(string)을 의미한다.
- \%c : 한 문자(a single character)를 의미한다.
- \%d : 정수(integer)를 의미한다.
- \%f : 실수(floating-point)를 의미한다.

이 포맷팅 방법은 현재 deprecated 상태이다. 즉, 중요도가 떨어져 더 이상 사용하지 않고 앞으로 사라지게 될 기능을 말한다. 그러나 오래전부터 코딩을 해오던 개발자들 일부는 이 포맷팅 방법을 여전히 사용하고 있기도 한다.

여러분들에게 이 포맷팅 방법을 사용하라고 권장하지는 않지만 이와 같은 코드를 만났을 때 무슨 뜻인지 이해는 할 수 있으면 도움이 될 것 같다.


- 문자열안에 %d와 문자열 뒤의 %3이 서로 매핑된다. 숫자 3이 들어갈 자리에 %d를 써준다.
- \% 뒤 데이터 자리에 변수가 올수도 있다.

````python
b = "I eat %d apples." % 3
print(b)
````

````python
number=3
"I eat %d apples." % number
````

- 문자열 'five'가 들어갈 자리에 %s를 써준다.

````python
"I eat %s apples." % "five"
````

- 변수가 2개 이상일 경우에는 소괄호() 즉, 튜플로 묶어준다.
- 아래 코드에서 차례대로 변수 number는 %d 자리에, day는 %s 자리에 들어간다.

````python
number = 10
day = "three"
"I ate %d apples. so I was sick for %s days." % (number, day)
````

- 숫자 3은 정수데이터이다. 그러나 숫자3이 들어갈 자리에 문자열 형식지정자인 %s를 써도 에러가 나지 않는다. 다른 프로그래밍 언어에서는 에러가 나지만 파이썬에서는 의미가 없어졌다.
- 즉, 정수 3이 문자열 '3'으로 변환되어 출력된다.

````python
"I have %s apples" % 3
````

- 실수 3.234가 실수형 형식지정자인 %f와 매핑되어 소수점 6자리로 출력된다.

````python
"The rate is %f" % 3.234
````

- 지끔까지 봤던 것처럼 %는 특수기능을 가지고 있는 문자이다. 그럼 % 기호 자체를 출력하려면 어떻게 해야 할까?
- Escape characters에서 처럼 % 기호를 앞에 한번 더 붙여주면 된다.

````python
"The error rate is %d%%." % 98
````

## "New Style" String Formating
Old style 포맷팅은 데이터 타입에 맞게 형식지정자를 적어줘야 했다. format() 함수는 이러한 번거로움을 없앴다.
format() 함수의 형식은 다음과 같다.
```
문자열 { } 문자열.format(data)
```
format() 함수에서는 데이터가 들어갈 자리를 { }로 표시해 놓는다.


- { } 자리에 변수 name의 값이 들어간다.

````python
name = 'Bob'
'Hello, {}'.format(name)
````

- { } 자리에 정수 3이 들어간다.

````python
"I eat {} apples".format(3)
````

- { } 자리에 문자열 'five'가 들어간다.

````python
"I eat {} apples".format("five")
````

- 데이터가 여러개라면 format( ) 괄호안에 인자로 나열해준다.

````python
print('We are the {} who say "{}!"'.format('knights', 'Ni'))
````

- { }가 여러개 있는 경우 인덱스를 붙여주기도 한다.

````python
number = 10
day = "three"
"I ate {0} apples. so I was sick for {1} days.".format(number, day)
````

- 인덱스를 정수로 붙여주기도 하지만 키워드로 붙일 수도 있다.
- 아직 배우지 않았지만 number와 day는 키워드 인자(keyword arguments)이다. 이 내용은 함수파트에서 다룬다.

````python
"I ate {number} apples. so I was sick for {day} days.".format(number=10, day=3)
````

- 인덱스와 키워드 인자를 혼합하여 써도 된다.

````python
"I ate {0} apples. so I was sick for {day} days.".format(10, day=3)
````

### 문자열 정렬

- 인덱스에 대해 문자열을 지정된 길이로 만든 뒤 오른쪽으로 정렬하고 남는 공간을 공백으로 채운다.

```
'{인덱스:>길이}'.format(값)
```



````python
print("He paid {0:>10} dollars.".format(10))
print("He paid {0:>10} dollars.".format(100))
print("He paid {0:>10} dollars.".format(1000))
print("He paid {2:>10} dollars.".format(10000, 1000, 200000))
````

- 인덱스에 대해 문자열을 지정된 길이로 만든 뒤 왼쪽으로 정렬하고 남는 공간을 공백으로 채운다.

```
'{인덱스:<길이}'.format(값)
```

````python
print("He paid {0:<10} dollars.".format(10))
print("He paid {0:<10} dollars.".format(100))
print("He paid {0:<10} dollars.".format(1000))
print("He paid {0:<10} dollars.".format(10000))
````

- 인덱스에 대해 문자열을 지정된 길이로 만든 뒤 중앙으로 정렬하고 남는 공간을 공백으로 채운다.

```
'{인덱스:^길이}'.format(값)
```

````python
print("He paid {0:^10} dollars.".format(10))
print("He paid {0:^10} dollars.".format(100))
print("He paid {0:^10} dollars.".format(1000))
print("He paid {0:^10} dollars.".format(10000))
````

### 문자열 채우기

- 인덱스에 대해 문자열을 지정된 길이로 만든 뒤 왼쪽으로 정렬하고 남는 공간을 점으로 채운다.
- 점(.) 대신에 다른 대체 문자를 넣을 수 있다.

```
'{인덱스:.<길이}'.format(값)
```

````python
print("He paid {0:.<10} dollars.".format(10))
print("He paid {0:.<10} dollars.".format(100))
print("He paid {0:.<10} dollars.".format(1000))
print("He paid {0:.<10} dollars.".format(10000))
````

### 실수 표현
- 모듈 math에 변수 pi를 이용하여 실수값에 대해 포맷팅을 해보자.

````python
import math
print(math.pi)
````

- 인덱스에 대해 총자리수 가운데 소수점이하 자리수로 표현한다.

```
'{인덱스:총자리수.소수점자리수}'.format(값)
```

````python
print("He paid {0:0.2f} dollars.".format(math.pi))
print("He paid {0:0.3f} dollars.".format(math.pi))
print("He paid {0:0.4f} dollars.".format(math.pi))
print("He paid {0:0.5f} dollars.".format(math.pi))
````

````python
print("He paid {0:10.2f} dollars.".format(math.pi))
print("He paid {0:10.3f} dollars.".format(math.pi))
print("He paid {0:10.4f} dollars.".format(math.pi))
print("He paid {0:10.5f} dollars.".format(math.pi))
````

````python
print("He paid {0:<10.2f} dollars.".format(math.pi))
print("He paid {0:<10.3f} dollars.".format(math.pi))
print("He paid {0:<10.4f} dollars.".format(math.pi))
print("He paid {0:<10.5f} dollars.".format(math.pi))
````

- 문자열 속에 들어갈 값이 여러 개일 경우 변수명과 키워드인자를 동일하게 주기도 한다.
- 예를 들어, 오른쪽 name은 키워드 인자, 왼쪽 name은 변수명이다.


```
.format(name=name,error=error)
```



````python
name = 'Bob'
error = 98
'Hey {name}, there is a {error}% error.'.format(name=name,error=error)
````

## String Interpolation (f-strings)

- 변수명과 키워드인자를 동일하게 쓰는 번거로움(.format(name=name))을 없애고 문자열의 중괄호안에 {변수명 또는 연산식}을 직접 쓴다.

````python
name = 'Bob'
f'Hello, {name}!'
````

````python
a = 5
b = 10
f'Five plus ten is {a + b} and not {2 * (a + b)}.'
````

- 딕셔너리도 가능하다.

````python
d = {'name':'홍길동', 'age':30}
f'나의 이름은 {d["name"]}입니다. 나이는 {d["age"]}입니다.'
````

````python
name = 'Bob'
error = 98
f'Hey {name}, there is a {error}% error.'
````

````python
f'{math.pi:10.4f}'
````

````python
f'{math.pi:10.5}'
````

- 별다른 지시가 없으면 기본적으로 문자열은 왼쪽 정렬, 숫자는 오른쪽 정렬된다.

````python
table = {'Kim': 4127, 'Byun': 4785, 'Kang': 7678}
for name, phone in table.items():
  print(f'{name:10} ==> {phone:10f}')
````
